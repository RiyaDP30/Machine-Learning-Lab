```python
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)

from sklearn.neural_network import MLPClassifier
import numpy as np

X = np.random.rand(100, 10)
y = np.random.randint(0, 2, 100)

model = MLPClassifier(hidden_layer_sizes=(10, 5), activation='relu',
                      solver='adam', max_iter=2000, random_state=42)
model.fit(X, y)

print("Model: 'MLPClassifier (Simulated ANN)'")
print("="*50)
print(f"{'Layer (type)':<25}{'Output Shape':<20}{'Param #'}")
print("="*50)
print(f"{'dense_1 (Dense)':<25}{'(None, 10)':<20}{(10*10 + 10)}")
print(f"{'dense_2 (Dense)':<25}{'(None, 5)':<20}{(10*5 + 5)}")
print(f"{'dense_3 (Dense)':<25}{'(None, 1)':<20}{(5*1 + 1)}")
print("="*50)
print(f"Total params: {(10*10+10) + (10*5+5) + (5*1+1)}")
print(f"Trainable params: {(10*10+10) + (10*5+5) + (5*1+1)}")
print(f"Non-trainable params: 0")
print("="*50)
```
